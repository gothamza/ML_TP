```python
m=data['Close'].resample('w').agg(['mean','std','min','max'])
m['mean']['2022'].plot(label='moyenne par  somaine')

plt.fill_between(m.index,m['max'],m['min'],label='min max pare semaine')
plt.legend()
plt.show
```

```python
m = data["Close"].resample('W').agg(["mean", "std", "min", "max"])
m["mean"]["2022"].plot()
plt.fill_between(m.index, m["max"], m["min"], alpha=0.2)
```

```python
data.loc['2022','Close'].rolling(window=7).mean().plot()
```

```python
plt.figure(figsize=(12,8))
data.loc['2022-6','Close'].plot()
data.loc['2022-6','Close'].rolling(window=7).mean().plot(label="non center")
data.loc['2022-6','Close'].rolling(window=7,center=True).mean().plot(label=" center")
data.loc['2022-6','Close'].ewm(alpha=0.6).mean().plot(label="owm center")
plt.legend()
plt.show
```

```python
plt.figure(figsize=(12,8))
data.loc['2022-6','Close'].plot()
data.loc['2022-6','Close'].rolling(window=7).mean().plot(label="non center")
data.loc['2022-6','Close'].rolling(window=7,center=True).mean().plot(label=" center")
for i in np.arange(0.2,1.2,0.2):
    data.loc['2022-6','Close'].ewm(alpha=i).mean().plot(label=f"owm center n{i}",linestyle='--')
plt.legend()
plt.show   
```
